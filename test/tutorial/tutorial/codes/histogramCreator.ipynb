```python
#Import packages
import pandas as pd
from pandas import DataFrame
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
filePath = '/Users/luisricardoferraz/book-similarity/test/tutorial/tutorial/spiders/clustering/'
```

```python
files = [{"file": 'log-AffinityPropagation.txt', "name": 'Affinity Propagation'}, {"file":'log-average-linkage.txt', "name": 'Average Linkage'}, {"file":'log-complete-linkage.txt', "name": 'Complete Linkage'}, {"file":'log-single-linkage.txt', "name": 'Single Linkage'}, {"file":'log-SpectralClustering.txt', "name": 'Spectral Clustering'}]
                                                                                                                
```

```python
for algorithm in files:
    myarray = []
    with open(filePath+algorithm.get("file"),'r') as file:
        for line in file:
            value = re.search('Total de livros do cluster [0-9]+: ([0-9]+)', line)
            if value != None:
                myarray.append(value[1])
            else:
                continue

    new_array = list(map(int, myarray))
    new_array.sort()
    my_set = list(set(new_array))
    my_set.sort()
    #print(my_set)
    #plt.figure(figsize=(16, 6))
    #plt.hist(new_array, bins=range(len(my_set)+1))
    #plt.title('Distribuição dos grupos pelo algoritmo '+algorithm.get("name"))
    #plt.xlabel('Tamanho do grupo')
    #plt.ylabel('Frequência absoluta')
    #plt.show()
    
    idade_sum = pd.Series(new_array).value_counts()
    idade_sum = idade_sum.sort_index()
    x = idade_sum.index
    y = idade_sum
    #print(x)
    #print(y)
    plt.figure(figsize=(16, 6))
    plt.bar(x, y)
    plt.title('Distribuição dos grupos pelo algoritmo '+algorithm.get("name"))
    plt.xlabel('Total de documentos no grupo')
    plt.ylabel('Frequência absoluta')
    plt.xticks(x)
    plt.savefig(filePath+'histograma-'+algorithm.get("name")+'.png')
    plt.show()
```
