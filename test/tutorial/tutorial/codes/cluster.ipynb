```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

import pandas as pd
from pandas import DataFrame
import numpy as np

from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import pairwise_distances
import pylab as pl

filePath = '/Users/luisricardoferraz/book-similarity/test/tutorial/tutorial/spiders/synopsis-preprocessing/'

X = pd.read_csv(filePath+'MatrixOfVectorizedSynopsis.csv')
titulos = X['titulo']
X = X.drop(['Unnamed: 0','titulo'],1)
```

```python
Y = X[:100]
Y
```

```python
num_clusters = 5
km = KMeans(n_clusters=num_clusters)
km.fit_predict(Y)
clusters = km.labels_.tolist()
print("Total de iterações:")
print(km.n_iter_)
print()
print("Distribuição dos clusters:")
print(clusters)
```

```python
def assignClustersToTitles(clusters,titulos):
    dataframe = pd.DataFrame()
    dataframe['titulo'] = titulos
    dataframe['cluster'] = clusters
    return dataframe
```

```python
def printClusters(num_clusters, dataframe):
    for i in range(num_clusters):
        new_dataframe = dataframe.loc[dataframe['cluster'] == i]
        lista_livros = []
        lista_livros = list(new_dataframe['titulo'])
        print('Total de livros do cluster %d' %i + ": " + str(len(lista_livros)) + "\n")
        print('Lista de livros do cluster %d' %i + ": " + ", ".join(lista_livros) + "\n")
```

```python
def extractMetrics(algorithm, dataset):
    if hasattr(algorithm, 'silhouette_score'):
        print("Quanto mais próximo de 1, melhor: ")
        print(metrics.silhouette_score(dataset, algorithm.labels_, metric='euclidean'))
        print("\n")
    if hasattr(algorithm, 'davies_bouldin_score'):
        print("Quanto menor, melhor: ")
        print(metrics.davies_bouldin_score(dataset, algorithm.labels_))
        print("\n")
    if hasattr(algorithm, 'inertia_'):
        print("Inércia: ")
        print(algorithm.inertia_)
        print("\n")
    if hasattr(algorithm, 'score'):
        print("Score: ")
        print(algorithm.score(Y))
```

```python
def reduceDimensionalityAndPlotFigure(dataset, clusters, title):
    pca = PCA(n_components=2).fit_transform(dataset)
    pl.figure('Reference Plot')
    pl.scatter(pca[:, 0], pca[:, 1], c=clusters)
    pl.title(title)
    pl.show()
```

```python
dataframe = assignClustersToTitles(clusters,titulos[:100])
printClusters(num_clusters, dataframe)
print("===================================")
extractMetrics(km, Y)
reduceDimensionalityAndPlotFigure(Y, dataframe['cluster'], 'K-means aplicado ao dataset')
```

```python
from sklearn.cluster import MiniBatchKMeans

minibkm = MiniBatchKMeans(n_clusters=num_clusters)
minibkm.fit_predict(Y)
minibclusters = minibkm.labels_.tolist()
print("Total de iterações:")
print(minibkm.n_iter_)
print()
print("Distribuição dos clusters:")
print(minibclusters)
```

```python
minibdataframe = assignClustersToTitles(minibclusters,titulos[:100])
printClusters(num_clusters, minibdataframe)
print("===================================")
extractMetrics(minibkm, Y)
reduceDimensionalityAndPlotFigure(Y, minibdataframe['cluster'], 'MiniBatchKmeans aplicado ao dataset')
```
